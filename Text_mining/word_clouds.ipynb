```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import PyPDF2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pdfFileObj = open(r'C:\Users\owner\Documents\adhoc\20190309_text_mining_source\2018-barclays-plc-annual-report.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
text=''
for pageNum in range(pdfReader.numPages):
    page_text=pdfReader.getPage(pageNum).extractText().replace('\n', '')
    text +=' '
    text +=page_text
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text.lower())

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text.lower())
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["Barclays", "Group", "PLC", "home", "barclays","Annual", "Report","year","annualreport","including"])
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text.lower())

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Generate a word cloud image
mask = np.array(Image.open(r'C:\Users\owner\Documents\adhoc\20190309_text_mining_source\Barclays-Logo-1.jpg'))
wordcloud= WordCloud(stopwords=stopwords, background_color="white", mode="RGBA", max_words=200, mask=mask).generate(text.lower())

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[40,40])
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

# store to file
plt.savefig(r'C:\Users\owner\Documents\adhoc\20190309_text_mining_source\Barclays.png', format="png")

plt.show()
```
